```python
import pymongo
import pandas as pd
import numpy as np
from src.metacritic_scrape import clean_title
```

```python
mc = pymongo.MongoClient()
db = mc['ps4_game_data']
games = db['games']
```

```python
def make_df(mongo=games):
    df = pd.DataFrame(list(mongo.find()))
    df = df.drop(df[df.category == 3].index)
    df = df.drop(df[df.category==2].index)
    df = df.drop(df[df.category==1].index)
    df = (df[df.version_parent.isnull()])
    return df
```

```python
df = make_df()
```

```python
lots = list(df.name)
```

```python
bar=[]
```

```python
for game in lots:
    bar.append(clean_title(game))
```

```python
for i in range(len(bar)):
    print(lots[i])
    print(bar[i])
```
