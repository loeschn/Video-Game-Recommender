The mongodb was incorrectly named ps4_game_data early on in my project, where I should have just named it game_data, as I am now adding other console platforms to the database

```python
mc=pymongo.MongoClient()
db=mc['ps4_game_data']
```

```python
platforms=db['platform']
games=db['games']
```

```python
platform_ids=list(range(150))
```

Example of what one platform endpoint looks like

```python
mac = req.api_endpoint(endpoint="platforms", endpoint_id="14")
```

Updating platforms db with all platforms contained in IGDB

```python
for platform_id in platform_ids:
    platform_data = req.api_endpoint(endpoint="platforms", endpoint_id=platform_id)
    if platform_data:
        platforms.insert_many(platform_data)
```

Adding switch games to games db

```python
switch = platforms.find_one({'name': 'Nintendo Switch'})
```

```python
switch_game_ids = switch['games']
```

```python
switch_game_ids
```

```python
len(switch_game_ids)
```

Finding out how many games from the Nintendo Switch platform already exists in my games db from just ps4 games

```python
yes=0
no=0
for game_id in switch_game_ids:
    if games.find_one({'id': game_id}):
        yes+=1
    else:
        no+=1
```

```python
yes, no
```

```python
for game_id in switch_game_ids:
    game_data = req.api_endpoint(endpoint="games", endpoint_id=game_id)
    if games.find_one({'id': game_id}):
        continue
    else:
        games.insert_many(game_data)
```
